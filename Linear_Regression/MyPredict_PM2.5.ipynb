```python
#首先导入需要的库
import numpy as np
import pandas as pd
```

```python
#读取数据文件
df = pd.read_csv('train.csv')
df.head(10)
```

如图可以看出，训练集的数据是各种观测值，时间为14年1月1日，包含PM2.5，接下来我们抽取需要训练的PM2.5

```python
df = df[df['observation']=='PM2.5']
df
```

```python
PMdata = df.iloc[:,3:]  #提取数值信息
PMdata.head(10)
```

接下来我们开始训练数据,9个9个一组训练，如上共有24列，所以循环15次，y是最后一列

```python
train_x=[]
train_y=[]
for i in range(15):
    x = PMdata.iloc[:,i:i+9]
    x.columns = np.array(range(9))
    y = PMdata.iloc[:,i+9]
    y.columns = np.array(range(1))
    train_x.append(x)
    train_y.append(y)
train_x=pd.concat(train_x)
train_y=pd.concat(train_y) 
train_x
```

```python
train_x=np.array(train_x,float)   #转化为小数
train_y=np.array(train_y,float)
train_x
```

给训练数据增加bias

```python
train_x=np.concatenate((np.ones((train_x.shape[0],1)),train_x),axis=1)
train_x
```

初始化权重并使用ada梯度下降训练数据

```python
w = np.zeros(len(train_x[0]))
lr = 0.1
Iteration = 10000
sum_gra=np.zeros(len(train_x[0])) 
for i in range(Iteration):
    y_new = np.dot(train_x,w)
    loss = y_new - train_y
    gra = 2*np.dot(train_x.transpose(),loss)
    sum_gra += gra**2
    ada = np.sqrt(sum_gra)
    w = w - lr*gra/ada
w
```

```python
df_test=pd.read_csv('test.csv')
df_test.head()
```

```python
df2=df_test[df_test['AMB_TEMP']=='PM2.5']
df2.head(10)
```

```python
PMdata_test=df2.iloc[:,2:]
PMdata_test
```

```python
test_x=np.array(PMdata_test,float)
test_x
```

```python
test_x=np.concatenate((np.ones((test_x.shape[0],1)),test_x),axis=1)
test_x
```

```python
y_pre=np.dot(test_x,w)
y_pre
```

```python
#将预测值写入csv文件
y_submit=pd.read_csv('sampleSubmission.csv')
y_submit.value=y_pre   #get value by .value   automatically identify numbers
y_submit.to_csv('sampleSubmission.csv',index=False)
```

```python
#计算误差
real=pd.read_csv('answer.csv')
error=abs(y_submit.value-real.value).sum()/len(real.value)
error
```
